```python
def sample_level(levels):
    new_level1 = []
    new_level2 = []
    new_level3 = []
    new_level4 = []

    for level in levels:
        level_copy = level.copy()

        # print(level_copy)

        random.shuffle(level_copy)
        
        new_level1.append(level_copy[0])
        new_level2.append(level_copy[1])
        new_level3.append(level_copy[2])
        new_level4.append(level_copy[3])

    return new_level1, new_level2, new_level3, new_level4
```

```python
stimuli_set = []
for i in range(5):
    l1 = new_levels[0]
    l2 = new_levels[1]

    random.shuffle(l1)
    random.shuffle(l2)


    new_stimuli1 = []
    new_stimuli2 = []
    new_stimuli3 = []
    new_stimuli4 = []
    
    # print(l1)
    # print(l2)
    new_stimuli1.append(l1[0])
    new_stimuli1.append(l2[0])

    new_stimuli2.append(l1[1])
    new_stimuli2.append(l2[1])

    new_stimuli3.append(l1[2])
    new_stimuli3.append(l2[2])

    new_stimuli4.append(l1[3])
    new_stimuli4.append(l2[3])

    new_level1, new_level2, new_level3, new_level4 = sample_level(new_levels[2:])

    new_stimuli1.extend(new_level1)
    new_stimuli2.extend(new_level2)
    new_stimuli3.extend(new_level3)
    new_stimuli4.extend(new_level4)

    random.shuffle(new_stimuli1)
    random.shuffle(new_stimuli2)
    random.shuffle(new_stimuli3)
    random.shuffle(new_stimuli4)

    stimuli_set.append(new_stimuli1)
    stimuli_set.append(new_stimuli2)
    stimuli_set.append(new_stimuli3)
    stimuli_set.append(new_stimuli4)
```
